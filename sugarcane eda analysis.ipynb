# Data Cleaning

```python
df.info()
```

```python
#notice that in production,production per year,acreage,yield have some mistake of decimal and comma


#common mistake-don't forget to put the comma in the df['<column name>']
```

```python
df['Production (Tons)']=df['Production (Tons)'].str.replace(".","")
df['Production per Person (Kg)']=df['Production per Person (Kg)'].str.replace(".","").str.replace(",",".")
df['Acreage (Hectare)']=df['Acreage (Hectare)'].str.replace(".","")
df['Yield (Kg / Hectare)']=df['Yield (Kg / Hectare)'].str.replace(".","").str.replace(",",".")
```

```python
df.head(5)
```

```python
df.isna().sum()
```

```python
df[df["Acreage (Hectare)"].isnull()]
```

```python
df[df["Yield (Kg / Hectare)"].isnull()]
```

```python
#means both are in one same row
df=df.dropna().reset_index()
df.head()
```

```python
df.isnull().sum()
```

```python
#dropping unnecessay column like unamed and index
df.drop(['Unnamed: 0','index'],axis=1,inplace=True)
```

```python
df.head()
```

```python
df.dtypes
```

```python
df['Production (Tons)']=df['Production (Tons)'].astype('float')
df['Production per Person (Kg)']=df['Production per Person (Kg)'].astype('float')
df['Acreage (Hectare)']=df['Acreage (Hectare)'].astype('float')
df['Yield (Kg / Hectare)']=df['Yield (Kg / Hectare)'].astype('float')
```
